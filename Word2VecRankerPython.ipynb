```python
import re
from nltk.corpus import stopwords
import gensim
from gensim.models import Word2Vec
import numpy as np
import glob
import nltk
```

```python
def get_embeddings(word):
    try:
        return w2vec[word]
    except:
        return np.zeros(300)
    
    
def get_similarity(query, doc):
    cos_sim = np.dot(query, doc)/(np.linalg.norm(query)*np.linalg.norm(doc))
    return cos_sim

def getNextBatch(file):
    batch = ""
    line = file.readline()
    batch += line.replace("Document", "")
    line = file.readline()
    while line:
        if ("****" in line):
            break
        batch += line 
        line = file.readline()
    return batch
    
     
def getNextQuery(file):
    batch = ""
    line = file.readline()
    while line:
        batch += line + " "
        if ("#" in line):
            batch = batch[0:-2]
            break       
        line = file.readline()
    return batch
    
def getAllDocs():
    filesList = sorted(glob.glob("lisa/*"))
    docs = []
    for fileName in filesList:
        if fileName =="lisa/LISA.QUE" or fileName =="lisa/LISA.REL" or fileName =="lisa/LISARJ.NUM" or fileName =="lisa/README" or fileName=="lisa/Merged" :
            continue
        file = open(fileName, 'r')
        current = getNextBatch(file)
        while current:
            docs.append( re.sub(r'[^A-Za ]+', ' ', current ))
            current = getNextBatch(file)
        file.close()
    return docs 
    

def getAllQueries():
    queries = []
    file = open("lisa/LISA.QUE", 'r')
    current = getNextQuery(file)
    while current:
        queries.append( re.sub(r'[^A-Za ]+', ' ', current ))
        current = getNextQuery(file)
    file.close()
    return queries


def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)
    text = re.sub(r'[^\w\s]+',' ', text)
   
    return text

def rank_docs(query):
    query_vector = np.mean(np.array([get_embeddings(word) for word in nltk.word_tokenize(preprocess(query))]),axis=0)
    rank = []
    for i,j in docVectors.items():
        k = list(j.keys())[0]
        v = list(j.values())[0]
        rank.append((k,get_similarity(query_vector, v)))
    
    rank = np.array([x[1] for x in rank])
    rankSims = rank.argsort()[::-1]
    
    return rankSims.tolist(), sorted(rank, reverse=True)
```

```python

docVectors = {}

i=0
for sent in getAllDocs():
    i+=1
    print(str(i) + "/6004 done")
    average_vector = np.mean(np.array([get_embeddings(word) for word in nltk.word_tokenize(preprocess(sent))]),axis=0)
    d = {i:{sent:average_vector}}
    docVectors.update(d)
```
